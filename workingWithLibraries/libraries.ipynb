### NumPy

```python
import numpy as np

# Create a 1D NumPy array
arr1 = np.array([1, 2, 3, 4, 5])
print(arr1)

# Create a 2D NumPy array
arr2 = np.array([[1, 2, 3], [4, 5, 6]])
print(arr2)

# Create a 3D   or multidinensional NumPy array
arr3 = np.array([[[1,2,3,4],[10,20,30,40],[100,200,300,400]]])
print(arr3)

#Create an array of zeros
arr4 = np.zeros((3,3))
print(arr4)

#to get the shape of an array
np.shape(arr3)

#to get the number of dimensions of an array
np.ndim(arr3)

#to get the number of elements in an array
np.size(arr3)

#to sort an array
np.sort(arr3)

#perfroms argmax on an array
np.argmax(arr3)

#slicing an array
print(arr3[0,1,2])

#statistical operations on an array
print(np.mean(arr3))
print(np.median(arr3))
print(np.var(arr3))

# Matrix multiplication
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])

print(np.dot(A, B))  

```

```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y1 = [1, 4, 9, 16, 25]
y2 = [25, 16, 9, 4, 1]

# Plotting
plt.plot(x, y1, label='y = x^2', color='blue')
plt.plot(x, y2, label='y = 25 - x^2', color='red')

# Adding title and labels
plt.title('Multiple Line Plot with Legend')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Show legend
plt.legend()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

# Line Plot with Customizations
plt.plot(x, y, color='purple', marker='o', linestyle='-', linewidth=2, markersize=8)
plt.title('Customized Line Plot', fontsize=14, color='black')
plt.xlabel('X-axis', fontsize=12, color='blue')
plt.ylabel('Y-axis', fontsize=12, color='green')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]
z = [25, 16, 9, 4, 1]

# Create subplots (1 row, 2 columns)
fig, axes = plt.subplots(1, 2)

# Plot in the first subplot
axes[0].plot(x, y, label='y = x^2', color='blue')
axes[0].set_title('Plot 1: y = x^2')

# Plot in the second subplots
axes[1].plot(x, z, label='z = 25 - x^2', color='red')
axes[1].set_title('Plot 2: z = 25 - x^2')

# Display the plots
plt.tight_layout()  # To avoid overlap
plt.show()
```
